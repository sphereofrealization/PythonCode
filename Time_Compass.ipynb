```python
import ipywidgets as widgets
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt

# Constants
R = 5  # Radius of the circle
omega = 2 * np.pi  # Angular velocity (in radians per second)
t_max = 10  # Maximum time

# Create sliders for time, radius and angular velocity
t_slider = widgets.FloatSlider(value=0, min=0, max=t_max, step=0.1, description="Time (s)")
R_slider = widgets.FloatSlider(value=R, min=1, max=10, step=0.5, description="Radius (m)")
omega_slider = widgets.FloatSlider(value=omega, min=0, max=4*np.pi, step=0.1*np.pi, description="Ang. Vel. (rad/s)")

# Define function to create the visualization
def visualize_circular_motion(t, R, omega):
    # Calculate position
    x = R * np.cos(omega * t)
    y = R * np.sin(omega * t)

    # Create figure and axes
    fig, ax = plt.subplots()

    # Draw circle
    circle = plt.Circle((0, 0), R, color='blue', fill=False)
    ax.add_artist(circle)

    # Draw path up to current time
    t_points = np.linspace(0, t, 100)
    x_path = R * np.cos(omega * t_points)
    y_path = R * np.sin(omega * t_points)
    ax.plot(x_path, y_path, color='gray', linestyle='--')

    # Draw arrow representing the particle
    ax.arrow(0, 0, x, y, head_width=R/20, head_length=R/15, fc='red', ec='red')

    # Scatter for the particle
    ax.scatter(x, y, color='red')

    # Labels and limits
    ax.set_xlabel("x (m)")
    ax.set_ylabel("y (m)")
    ax.set_xlim(-R * 1.5, R * 1.5)
    ax.set_ylim(-R * 1.5, R * 1.5)
    ax.set_aspect('equal')

    # Display the plot
    plt.show()

# Link sliders to visualization function
widgets.interact(visualize_circular_motion, t=t_slider, R=R_slider, omega=omega_slider)
```
