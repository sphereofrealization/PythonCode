```python
import numpy as np

import random

import matplotlib.pyplot as plt



# Constants and Parameters

rho = 0.6

mu = 1.0

K = 5

delta = 0.1

K_max = 10

eta = 0.5

t1 = 0

t_max = 100



# Utility Functions

def P_star(u_i, t):

    return random.uniform(0, 1)



def BP_star(u_i, t):

    return random.uniform(0, 1)



def TGG_star(u_i, t):

    return random.uniform(0, 1)



def RGG_star(u_i, t):

    return random.uniform(0, 1)



def Gamma(n):

    return np.exp(-n)



def distortion(u_i, eta_i):

    return min(u_i, eta_i)



def U_initial(size):

    return np.zeros(size)



# Data collection for plotting

m_values = []

i_values = []

gamma_values = []

rate_values = []



# Main Algorithm Function

def task_scheduling_SRPT_round_robin_DEADLINE_DRIVEN_INTERACTIVE_TE(rho, mu, K, delta, K_max):

    U_0 = U_initial(1)



    for m in range(1, K_max + 1):

        u_list = range(1, m + 1)  # List of task indices



        for i in range(1, m + 1):

            u_i = u_list[(i - 1) * len(u_list) // m]

            eta_i = eta



            D_min = distortion(u_i, eta_i)

            D_max = distortion(u_i, eta_i)



            # Solve Distortion as a function of Rate

            D_star = distortion(u_i, eta_i)



            # Reduce rate using Random Early Detection Algorithm

            R_star = U_0 - sum(P_star(u_i, t) * BP_star(u_i, t) * TGG_star(u_i, t)

                              for t in range(t1, t_max - u_i + 1))



            # Given the amount of rate leftover, solve the problem using Gamma(n)

            leftover_rate = float(R_star)

            optimal_gamma = Gamma(len(u_list))



            # Discard the bandwidth overhead note

            RGG_sum = sum(P_star(u_i, t) * BP_star(u_i, t) * RGG_star(u_i, t)

                          for t in range(t1, t_max - u_i + 1))



            final_gamma = float(optimal_gamma)



            # Collect data for plotting

            m_values.append(m)

            i_values.append(i)

            gamma_values.append(final_gamma)

            rate_values.append(leftover_rate)



            print(f"Optimal Gamma for m={m}, i={i}: {final_gamma}")



# Run the Task Scheduling Algorithm

task_scheduling_SRPT_round_robin_DEADLINE_DRIVEN_INTERACTIVE_TE(rho, mu, K, delta, K_max)



# Plotting the Results

plt.figure(figsize=(12, 6))



# Subplot 1: Gamma values

plt.subplot(1, 2, 1)

plt.scatter(m_values, gamma_values, c='blue', label='Gamma Values')

plt.xlabel('m (Iterations)')

plt.ylabel('Optimal Gamma')

plt.title('Optimal Gamma Values vs Iterations')

plt.legend()



# Subplot 2: Rate values

plt.subplot(1, 2, 2)

plt.scatter(m_values, rate_values, c='green', label='Leftover Rate Values')

plt.xlabel('m (Iterations)')

plt.ylabel('Leftover Rate')

plt.title('Leftover Rate Values vs Iterations')

plt.legend()



plt.tight_layout()

plt.show()
```
