```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, fixed

# Define the function to plot in spherical coordinates
def func(beta, theta):
    return -(1.0 * np.sqrt(3.5481432270250993e18 -
                          1.1294090667581471e18 * theta +
                          8.987551787368176e16 * theta**2))/(np.sqrt(
                                                      39.47841760435743 -
                                                      12.566370614359172 * theta + (2 * (np.pi +
                                                                                       np.sqrt(np.pi**2 - np.pi**2 * np.sin(beta)**2)))**2))

# Create the spherical plot with interactive rotation
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

beta = np.linspace(-np.pi/2, np.pi/2, 100)
theta = np.linspace(-2*np.pi, 2*np.pi, 200)
beta, theta = np.meshgrid(beta, theta)

r = func(beta, theta)
x = r * np.sin(beta) * np.cos(theta)
y = r * np.sin(beta) * np.sin(theta)
z = r * np.cos(beta)

ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Create interactive rotation controls
def rotate_plot(elev=30, azim=-60):
    ax.view_init(elev, azim)
    plt.draw()

interact(rotate_plot, elev=(-90, 90), azim=(-180, 180))
plt.show()
```
