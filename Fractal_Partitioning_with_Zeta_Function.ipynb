```python
import mpmath
import numpy as np
import matplotlib.pyplot as plt

# Initialize mpmath to higher precision
mpmath.mp.dps = 25  # set decimal places precision

# Generate a list of non-trivial zeros of the zeta function
def generate_zeros(num_zeros):
    """
    Generate the first `num_zeros` non-trivial zeros of the Riemann zeta function.
    """
    zeros = [mpmath.zetazero(n) for n in range(1, num_zeros + 1)]
    return zeros

# Apply the fractal partitioning (simulated with subset checks)
def fractal_partitioning(zeros, num_partitions):
    """
    Partition the list of zeros into `num_partitions` parts and verify properties.
    """
    partitions = [zeros[i::num_partitions] for i in range(num_partitions)]
    return partitions

# Verify properties of each partition
def verify_partitions(partitions):
    """
    Verify each partition for given properties of zeros.
    """
    for i, partition in enumerate(partitions):
        print(f"Checking partition {i + 1}:")
        for zero in partition:
            real_part = mpmath.re(zero)
            if real_part != 0.5:
                print(f"Verification failed for zero: {zero} (Real part: {real_part})")
                return False
        print(f"Partition {i + 1} verified successfully.")
    return True

# Compute the Riemann zeta function at a given complex number
def evaluate_zeta(s):
    """
    Compute the Riemann zeta function at a given complex number `s`.
    """
    return mpmath.zeta(s)

# Visualize the zeros on the complex plane
def visualize_zeros(zeros):
    re_parts = [mpmath.re(zero) for zero in zeros]
    im_parts = [mpmath.im(zero) for zero in zeros]

    plt.scatter(re_parts, im_parts, color='blue')
    plt.axvline(x=0.5, color='red', linestyle='--')
    plt.title('Non-trivial Zeros of the Riemann Zeta Function')
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.grid(True)
    plt.show()

# Main function to demonstrate the process
def main():
    num_zeros = 100  # Number of non-trivial zeros to consider
    num_partitions = 5  # Number of partitions for fractal analysis

    # Generate zeros
    zeros = generate_zeros(num_zeros)
    print(f"Generated {num_zeros} non-trivial zeros of the Riemann zeta function.")

    # Visualize zeros
    visualize_zeros(zeros)

    # Apply fractal partitioning
    partitions = fractal_partitioning(zeros, num_partitions)

    # Verify partitions
    if verify_partitions(partitions):
        print("All partitions verified successfully to have real part 0.5.")
    else:
        print("Verification failed.")

    # Demonstrate evaluation of the zeta function at specific points
    test_points = [0.5 + 14.135j, 0.5 + 21.022j, 0.5 + 25.011j]  # Known non-trivial zeros
    for point in test_points:
        zeta_value = evaluate_zeta(point)
        print(f"Zeta({point}) = {zeta_value}")

if __name__ == "__main__":
    main()
```
