```python
import matplotlib.pyplot as plt

# Setup the figure and axis
fig, ax = plt.subplots(figsize=(10, 10))

# Hide the axes
ax.axis('off')

# Position of each label
positions = {
    '0': (0.1, 0.9),
    'X': (0.9, 0.9),
    'C1': (0.5, 0.9),
    'A': (0.1, 0.7),
    'T(A)': (0.9, 0.7),
    'C2': (0.5, 0.7),
    'C3': (0.5, 0.5),
    'T(C)': (0.9, 0.5),
    'C4': (0.1, 0.3)
}

# Add text (labels)
for label, (x, y) in positions.items():
    ax.text(x, y, label, ha='center', va='center')

# Adding arrows
def draw_arrow(start, end, text="", transform=False):
    ax.annotate(
        text,
        xy=positions[start], xycoords='data',
        xytext=positions[end], textcoords='data',
        arrowprops=dict(arrowstyle="->", lw=1.5, color='black')
    )
    if transform:
        ax.annotate(
            transform,
            xy=((positions[start][0]+positions[end][0])/2, (positions[start][1]+positions[end][1])/2),
            xycoords='data',
            textcoords='data',
            ha='center', va='center'
        )

arrows = [
    ('0', 'C1', r'$\psi \oplus 1_X$'),
    ('C1', 'A', r'$\alpha \oplus \beta$'),
    ('A', 'C2', r'$p_1 \oplus p_2$'),
    ('C2', 'T(A)', r'$f \circ \psi$'),
    ('C2', 'C3', ''),
    ('C3', 'T(C)', r'$f$'),
    ('T(C)', 'T(A)', r'$f \circ h$'),
    ('C3', 'C4', r'$h$')
]

# Draw each arrow
for arrow in arrows:
    draw_arrow(*arrow)

# Add the unique transformations
transformations = [
    ('C1', 'X', r'$1_X$'),
    ('C2', 'C3', r'$\chi$'),
    ('C3', 'C4', r'$f$'),
    ('C4', 'C1', r'$\psi$')
]

for arrow in transformations:
    draw_arrow(*arrow, transform=arrow[2])

plt.show()
```
