```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource

def fractal_morphism(x, y, iterations=4, scale=0.5):
    """
    Apply a fractal transformation to a 2D grid.

    :param x, y: 2D grids of coordinates
    :param iterations: Number of fractal iterations
    :param scale: Scale factor for each iteration
    :return: Transformed height map
    """
    height = np.zeros_like(x)
    for i in range(iterations):
        # This is a very simple fractal formula, similar to diamond-square algorithm
        height += np.sin(x * (i+1) * 0.1) * np.cos(y * (i+1) * 0.1) * scale**(i+1)

    return height

# Set up the grid
nx, ny = 100, 100
x = np.linspace(0, 10, nx)
y = np.linspace(0, 10, ny)
X, Y = np.meshgrid(x, y)

# Apply fractal morphism
Z = fractal_morphism(X, Y)

# Create a light source for better 3D visualization
ls = LightSource(azdeg=315, altdeg=45)

# Plot the original and morphed worldsheet
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6), subplot_kw={'projection': '3d'})

# Original flat worldsheet
ax1.plot_surface(X, Y, np.zeros_like(X), cmap='viridis', edgecolor='none')
ax1.set_title('Original 2D String Worldsheet')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Fractal morphed worldsheet
rgb = ls.shade(Z, cmap=plt.get_cmap('viridis'), vert_exag=5, blend_mode='soft')
ax2.plot_surface(X, Y, Z, facecolors=rgb, cmap='viridis', edgecolor='none')
ax2.set_title('Fractal Morphism of the Worldsheet')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource

def enhanced_fractal_morphism(x, y, iterations=5, scale=0.5, frequency=0.2, phase_shift=0.1):
    """
    Apply an enhanced fractal transformation to a 2D grid.

    :param x, y: 2D grids of coordinates
    :param iterations: Number of fractal iterations
    :param scale: Scale factor for each iteration
    :param frequency: Frequency multiplier for increased detail
    :param phase_shift: Phase shift for adding more variation in pattern
    :return: Transformed height map
    """
    height = np.zeros_like(x)
    for i in range(iterations):
        angle = (i + 1) * frequency
        phase = (i + 1) * phase_shift
        # Enhance fractal by adding phase shift and tweaking frequency
        height += np.sin(x * angle + phase) * np.cos(y * angle + phase) * scale**(i+1)

    return height

# Set up the grid
nx, ny = 200, 200  # Larger grid for higher resolution
x = np.linspace(0, 10, nx)
y = np.linspace(0, 10, ny)
X, Y = np.meshgrid(x, y)

# Apply enhanced fractal morphism
Z = enhanced_fractal_morphism(X, Y, iterations=6, frequency=0.3)

# Create a light source for more dynamic 3D visualization
ls = LightSource(azdeg=315, altdeg=45)

# Plot the original and morphed worldsheet
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6), subplot_kw={'projection': '3d'})

# Original flat worldsheet
ax1.plot_surface(X, Y, np.zeros_like(X), cmap='gray', edgecolor='none', alpha=0.5)
ax1.set_title('Original 2D String Worldsheet')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Enhanced fractal morphed worldsheet
terrain_cmap = plt.get_cmap('terrain')  # Get the colormap object
rgb = ls.shade(Z, cmap=terrain_cmap, vert_exag=0.5, blend_mode='soft')
ax2.plot_surface(X, Y, Z, facecolors=rgb, edgecolor='none')
ax2.set_title('Enhanced Fractal Morphism of the Worldsheet')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit

@jit(nopython=True)
def compute_custom_fractal(real, imag, width, height, max_iter):
    fractal_set = np.zeros((height, width), dtype=np.float32)
    for i in range(height):
        for j in range(width):
            c = complex(real[j], imag[i])
            z = complex(0, 0)
            theta = np.angle(c)  # Use the angle of c as initial theta
            r = 0.1
            theta_history = [theta, theta]

            for k in range(max_iter):
                # More complex update for theta
                theta = (theta_history[0] * 0.9 + theta_history[1] * 0.1 +
                         0.1 * np.sin(k * 0.005) + np.cos(k * 0.001) * 0.1) % (2 * np.pi)
                theta_history = [theta, theta_history[0]]  # Shift history

                # Adjust growth rate of r
                r = 0.1 * (1 + np.sin(k * 0.001) * 0.5 + np.cos(k * 0.0005) * 0.1) ** (k / 1000)

                # Convert polar to Cartesian for z computation
                z_prev = z  # Save previous z for comparison
                z = r * np.exp(1j * theta) + c

                # Escape condition with smoothing
                if abs(z) > 2.0:  # Reduced from 4.0
                    # Smooth coloring based on the rate of divergence
                    smooth_color = k + 1 - np.log(np.log(abs(z))) / np.log(2)
                    fractal_set[i, j] = float(smooth_color)
                    break
                else:
                    # If z isn't escaping, use a recursive relationship
                    z = 0.5 * (z_prev + z)

    return fractal_set

def measure_activity(fractal_set):
    # Use average value for activity instead of standard deviation
    return np.mean(fractal_set)

def scan_fractal(x_min, x_max, y_min, y_max, resolution, subregion_size, max_iter):
    width = height = resolution
    real = np.linspace(x_min, x_max, width)
    imag = np.linspace(y_min, y_max, height)

    fractal_set = compute_custom_fractal(real, imag, width, height, max_iter)

    num_subregions = resolution // subregion_size
    activity_map = np.zeros((num_subregions, num_subregions))

    for i in range(num_subregions):
        for j in range(num_subregions):
            subregion = fractal_set[i*subregion_size:(i+1)*subregion_size, j*subregion_size:(j+1)*subregion_size]
            activity_map[i, j] = measure_activity(subregion)

    return activity_map, fractal_set

def plot_activity_map(activity_map, x_min, x_max, y_min, y_max):
    plt.imshow(activity_map, extent=[x_min, x_max, y_min, y_max], origin='lower', cmap='hot')
    plt.colorbar(label='Activity Measure (Mean)')
    plt.title('Activity Map of Fractal')
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.show()

def zoom_into_active_region(x_min, x_max, y_min, y_max, resolution, subregion_size, max_iter, num_zoom_levels):
    for level in range(num_zoom_levels):
        activity_map, fractal_set = scan_fractal(x_min, x_max, y_min, y_max, resolution, subregion_size, max_iter)
        plot_activity_map(activity_map, x_min, x_max, y_min, y_max)

        i, j = np.unravel_index(np.argmax(activity_map), activity_map.shape)
        subregion_width = (x_max - x_min) / (resolution // subregion_size)
        subregion_height = (y_max - y_min) / (resolution // subregion_size)

        x_min = x_min + j * subregion_width
        x_max = x_min + subregion_width
        y_min = y_min + i * subregion_height
        y_max = y_min + subregion_height

        title = f'Zoom Level {level + 1} - Fractal Exploration'
        explore_fractal(resolution, x_min, x_max, y_min, y_max, max_iter, title)

def explore_fractal(n, x_min, x_max, y_min, y_max, max_iter, title):
    real = np.linspace(x_min, x_max, n)
    imag = np.linspace(y_min, y_max, n)

    fractal_set = compute_custom_fractal(real, imag, n, n, max_iter)

    plt.figure(figsize=(10, 8))
    plt.imshow(fractal_set.T, extent=[x_min, x_max, y_min, y_max], cmap='viridis', origin='lower')
    plt.colorbar()
    plt.title(title)
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.show()

# Initial parameters for the broad view
x_min, x_max = -2, 2
y_min, y_max = -2, 2
resolution = 1000
subregion_size = 100
max_iter = 1000  # Increased iterations
num_zoom_levels = 3  # You can adjust the number of zoom levels

zoom_into_active_region(x_min, x_max, y_min, y_max, resolution, subregion_size, max_iter, num_zoom_levels)
```
